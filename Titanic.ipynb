```python
import pandas as pd
titanic = pd.read_csv('train.csv')
titanic.head(5)
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic.drop(['Name', 'PassengerId', 'Ticket', 'Cabin', 'Fare'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
titanic['Sex'] = encoder.fit_transform(titanic['Sex'])
titanic['Embarked'] = encoder.fit_transform(titanic['Embarked'])
```

```python
titanic
```

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
```

```python
import numpy as np
titanic['Age'] = titanic['Age'].apply(np.ceil).astype(int)
```

```python
titanic
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
data_scaled = ss.fit_transform(titanic.iloc[:,1:7]) 
print(data_scaled)
```

```python
titanic
```

```python
from sklearn.model_selection import train_test_split

X = titanic.iloc[:, 1:7] 
y = titanic['Survived']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 10)

X_train= ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
X_train
```

```python
#Setup arrays to store training and test accuracies
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
neighbors = np.arange(1, 10, 2)
train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)

    #Fit the model
    knn.fit(X_train, y_train)

    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)

    #Compute accuracy on the test set
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = KNeighborsClassifier(n_neighbors=11)
model.fit(X_train, y_train)
print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred= model.predict(X_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred=model.predict(X_test)))

# Predict on the test data to get y_pred with the correct size
y_pred = model.predict(X_test)

# Now generate the classification report
print(classification_report(y_test, y_pred))
```
